````
%%writefile dashboard.py
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

st.set_page_config(page_title="🚗 Cars Dashboard", layout="wide", page_icon="🚗")

st.markdown("""
<style>
    .main { background-color: #0f1117; }
    .block-container { padding-top: 1.5rem; }
    h1 { color: #00d4ff; font-size: 2.2rem; text-align: center; }
    .stMetric label { color: #aaa; }
</style>
""", unsafe_allow_html=True)

@st.cache_data
def load_data():
    df = pd.read_csv("cars_updated_2.csv")
    return df

df = load_data()

st.title("🚗 Cars Market Dashboard")
st.markdown("---")

# ── KPI Row ───────────────────────────────────────────
k1, k2, k3, k4 = st.columns(4)
k1.metric("Total Cars", f"{len(df):,}")
k2.metric("Brands", df['brand'].nunique())
k3.metric("Avg Price", f"${df['price'].mean():,.0f}")
k4.metric("Avg Horsepower", f"{df['horsepower'].mean():,.0f} HP")

st.markdown("---")

# ── Sidebar filter ─────────────────────────────────────
all_brands = sorted(df['brand'].unique())
selected_brands = st.sidebar.multiselect("Filter by Brand", all_brands, default=all_brands[:15])
if selected_brands:
    df = df[df['brand'].isin(selected_brands)]

# ═══════════════════════════════════════════════════════
# ROW 1  –  Sales Count per Brand per Year | Avg Mileage
# ═══════════════════════════════════════════════════════
st.subheader("📊 Sales Volume & Mileage Analysis")
col1, col2 = st.columns(2)

with col1:
    st.markdown("#### 🔢 Sales Count per Brand per Year")
    yearly = df.groupby(['brand', 'model_year']).size().reset_index(name='count')
    fig1 = px.bar(
        yearly, x='model_year', y='count', color='brand',
        labels={'model_year': 'Year', 'count': 'Sales Count', 'brand': 'Brand'},
        template='plotly_dark', height=430,
        color_discrete_sequence=px.colors.qualitative.Bold, barmode='stack'
    )
    fig1.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        legend=dict(orientation='v', x=1.01, y=1),
        margin=dict(l=10, r=10, t=10, b=10), xaxis=dict(tickangle=-45))
    st.plotly_chart(fig1, use_container_width=True)

with col2:
    st.markdown("#### 🛣️ Avg Mileage per Brand")
    avg_mil = df.groupby('brand')['milage'].mean().reset_index()
    avg_mil.columns = ['brand', 'avg_mileage']
    avg_mil = avg_mil.sort_values('avg_mileage', ascending=True)
    fig2 = px.bar(
        avg_mil, x='avg_mileage', y='brand', orientation='h',
        labels={'avg_mileage': 'Avg Mileage (miles)', 'brand': 'Brand'},
        template='plotly_dark', color='avg_mileage',
        color_continuous_scale='Blues', height=430,
    )
    fig2.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        coloraxis_showscale=False, margin=dict(l=10, r=10, t=10, b=10))
    st.plotly_chart(fig2, use_container_width=True)

# ═══════════════════════════════════════════════════════
# ROW 2  –  Top 10 Ext Colors | Top 10 Int Colors
# ═══════════════════════════════════════════════════════
st.markdown("---")
st.subheader("🎨 Color Popularity Analysis")
col3, col4 = st.columns(2)

with col3:
    st.markdown("#### 🚘 Top 10 Exterior Colors per Brand")
    ext_counts = df.groupby(['brand', 'ext_col']).size().reset_index(name='count')
    ext_top10 = (ext_counts.sort_values('count', ascending=False)
                 .groupby('brand').head(10).reset_index(drop=True))
    fig3 = px.bar(
        ext_top10, x='count', y='brand', color='ext_col', orientation='h',
        labels={'count': 'Count', 'brand': 'Brand', 'ext_col': 'Exterior Color'},
        template='plotly_dark', height=430,
        color_discrete_sequence=px.colors.qualitative.Pastel,
    )
    fig3.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        legend=dict(orientation='h', y=-0.25, x=0),
        margin=dict(l=10, r=10, t=10, b=60))
    st.plotly_chart(fig3, use_container_width=True)

with col4:
    st.markdown("#### 🪑 Top 10 Interior Colors per Brand")
    int_counts = df.groupby(['brand', 'int_col']).size().reset_index(name='count')
    int_top10 = (int_counts.sort_values('count', ascending=False)
                 .groupby('brand').head(10).reset_index(drop=True))
    fig4 = px.bar(
        int_top10, x='count', y='brand', color='int_col', orientation='h',
        labels={'count': 'Count', 'brand': 'Brand', 'int_col': 'Interior Color'},
        template='plotly_dark', height=430,
        color_discrete_sequence=px.colors.qualitative.Antique,
    )
    fig4.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        legend=dict(orientation='h', y=-0.25, x=0),
        margin=dict(l=10, r=10, t=10, b=60))
    st.plotly_chart(fig4, use_container_width=True)

# ═══════════════════════════════════════════════════════
# ROW 3  –  Avg Horsepower | Avg Price
# ═══════════════════════════════════════════════════════
st.markdown("---")
st.subheader("⚡ Performance & Pricing Analysis")
col5, col6 = st.columns(2)

with col5:
    st.markdown("#### ⚡ Avg Horsepower per Brand")
    avg_hp = df.groupby('brand')['horsepower'].mean().reset_index()
    avg_hp.columns = ['brand', 'avg_hp']
    avg_hp = avg_hp.sort_values('avg_hp', ascending=False)
    fig5 = px.bar(
        avg_hp, x='brand', y='avg_hp',
        labels={'avg_hp': 'Avg Horsepower (HP)', 'brand': 'Brand'},
        template='plotly_dark', color='avg_hp',
        color_continuous_scale='Oranges', height=420,
    )
    fig5.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        coloraxis_showscale=False, xaxis=dict(tickangle=-45),
        margin=dict(l=10, r=10, t=10, b=80))
    st.plotly_chart(fig5, use_container_width=True)

with col6:
    st.markdown("#### 💰 Avg Price per Brand")
    avg_price = df.groupby('brand')['price'].mean().reset_index()
    avg_price.columns = ['brand', 'avg_price']
    avg_price = avg_price.sort_values('avg_price', ascending=False)
    fig6 = px.bar(
        avg_price, x='brand', y='avg_price',
        labels={'avg_price': 'Avg Price (USD)', 'brand': 'Brand'},
        template='plotly_dark', color='avg_price',
        color_continuous_scale='Greens', height=420,
    )
    fig6.update_layout(plot_bgcolor='#1a1d2e', paper_bgcolor='#1a1d2e',
        coloraxis_showscale=False, xaxis=dict(tickangle=-45),
        margin=dict(l=10, r=10, t=10, b=80))
    fig6.update_yaxes(tickprefix='$', tickformat=',.0f')
    st.plotly_chart(fig6, use_container_width=True)

st.markdown("---")
st.caption("Cars Market Analysis Dashboard · Built with Streamlit & Plotly")
````

## ▶️ How to Run the Dashboard

**Option 1 – Local machine:**
```bash
# Make sure the CSV file is in the same folder as dashboard.py
streamlit run dashboard.py
```

**Option 2 – Inside Jupyter (using a background process):**
Run the cell below 👇


````
import subprocess, threading, time

def run_app():
    subprocess.run(["streamlit", "run", "dashboard.py",
                    "--server.port=8501", "--server.headless=true"])

t = threading.Thread(target=run_app, daemon=True)
t.start()
time.sleep(3)
print("✅ Dashboard running at: http://localhost:8501")
print("📌 Make sure cars_updated_2.csv is in the same directory as dashboard.py")
````
